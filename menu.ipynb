# Video Game Management System

This is a module that is the entry point of the CLI based video game management system.
The main_menu() allows the code to be directed into either renting, returning, searching, and inventory pruning the video games.

## Functions
- **draw_line()**: this function is to draw lines, length: length of the line, symbol: symbol used to create the line
- **display_message()**:this function is to make a banner for the Menu, message: displayed between the draw_lines, symbol: from draw_line function, space: the allign the message in the centre
- **display_menu()**: Used to display the prompt the user will read in the menu
- **main_menu()**: This is the CLI aspect of the menu allows user to choose between renting, returning, searching, and inventory pruning.

## Files 
- **database.py**: The contains the functions to check the customer ID, get the subscription tyoe, and get the customer's rental limit.
- **renting.py**: The contains the function that allows the customer to rent out a video game if it is available, and adds this as a new record to the Rental.txt
- **searching.py**: This contains the function that allows the customer to search for a given title, genre, or platform of a video game. Also shows game availability. 
- **returning.py**: This contains a function that allows the customer to return a video game they have rented previously. With the correct customer ID and its corresponding currently rented videogame, it adds the new record to the Rental.txt. Also asks for feedback, in tearms of ratings and comments, which are then saved to Game_Feedback.txt
- **inventorypruning.py**: This file contains the function that outputs the 10 most unpopular games and displays this data in a barchart.  The graph included in the inventory pruning function is a bar chart made using matplotlib.

```python
#Coursework Version 0.3
#Date: 05/12/2023
#By: F330446

'''
The CLI Module:
This module contains the 3 functions which all read through the text files and
output needed information for the main functions such as renting, searching to function.

- draw_line(): this function is to draw lines, length indicates of the line,symbol 
used to create the line.
- display_message():this function is to make a banner for the Menu, message is the 
string that will be displayed between the draw_lines. Symbol: from draw_line function.
Space: the allign the message in the centre
- display_menu(): This will display the options given to the user for the CLI.
- main_menu(): This is the CLI part, which will direct the user to the renting, searching,
returning, inventory pruning files. Or simply ends the program.
'''
#-------------------------------------
# Importing Files:
#-------------------------------------

import inventorypruning as IP
import renting as RG
import searching as SG
import returning as RTG

#-------------------------------------
# Function Definitions:
#-------------------------------------

def draw_line(length, symbol):
    line=length*symbol
    print(line)
    
def display_message(message):
    symbol="*"
    space=" "
    size=50
    length_message=len(message)
    side_length=round((size-length_message)/2)
    side=side_length*space
    draw_line(size,symbol)
    print(side+message+side)
    draw_line(size,symbol)

def display_menu():
    display_message("MENU")
    print("s) Search Game")
    print("r) Rent Game")
    print("t) Return Game")
    print("i) Inventory Pruning")
    print("e) Exit")
             
                
def main_menu():
    while True:
        display_menu()
        choice=input("Please enter your option(s/r/t/i/e): ").lower()
        if choice=="s":
            SG.search_game()
            display_message("END OR SEARCH")
        elif choice=="r":
            RG.rent_game()
            display_message("END OR RENT")
        elif choice=="t":
            RTG.return_game()
            display_message("END OF RETURN")
        elif choice=="i":
            IP.inv_pruning()
            display_message("END OF INVENTORY PRUNING")
        elif choice=="e":
            display_message("END")
        else:
            print("Sorry, invalid option. Please try again...")


# -----------------------------
# Main Program
# -----------------------------
main_menu()

#------------------------------
#Test Cases:
#------------------------------
'''
if __name__ == "__main__":

    # Test case to demonstrate the functions of draw_line():
    # Test Case 1: A line of 20 dots
    print("\n Test Case 1: A line of 20 dots")
    length.input="20"
    symbol.input="."
    draw_line(length, symbol)
    #Test case to demonstrate the function of display_message():
    #Test Case 2: A Successful message "Hello" outputted
    print("\n Test Case 2: A Successful message "Hello" outputted")
    message.input="Hello"
    display_message(message)

    # Test cases to demonstrate the functions of display_menu() and main_menu():
    # Test Case 3.1: A Successful input
    print("\n Test Case 3.1: A Successful input")
    choice.input = "s"
    SG.input="Here search function runs"
    main_menu()
    # Test Case 3.2: An Unsuccessful input
    print("\n Test Case 3.2: An Unsuccessful input")
    choice.input="f"
    main_menu()
    # Test Case 3.3: A Successful input- Uppercase
    print("\n Test Case 3.1: A Successful input- Uppercase")
    choice.input = "S"
    SG.input="Here search function runs"
    main_menu()

    #Resetting inputs to original state;
    length.input = input
    symbol.input= input
    message.input= input
    choice.input=input
    SG.input=input
'''    
```
